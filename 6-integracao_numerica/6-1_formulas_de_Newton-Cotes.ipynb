<a name="TFC">‚ö°</a> __Demonstra√ß√£o do primeiro teorema fundamental do C√°lculo:__

Isto mostra que
$$

üìù As f√≥rmulas de Newton-Cotes discutidas na $ \S 3 $ s√£o √†s vezes qualificadas como **simples**. Todas elas t√™m tamb√©m uma vers√£o composta em que $ N $ aparece como par√¢metro, obtida da maneira descrita acima.

# M√©todos de Taylor

## $ \S 1 $ Introdu√ß√£o

Seja $ \boldsymbol y $ a solu√ß√£o de um problema de valor inicial de primeira ordem em $ \mathbb R^r $:
\begin{equation*}\label{E:ivp}
\begin{cases}
\boldsymbol y' = \boldsymbol F(t, \boldsymbol y) \\
\boldsymbol y(t_0) = \boldsymbol y_0
\end{cases}
\quad \text{onde} \quad \boldsymbol y = \big(y_0, y_1, \dots, y_{r-1}\big)\,. \tag{1}
\end{equation*}

Usando o polin√¥mio de Taylor de ordem $ m \ge 1 $ de $ \boldsymbol y $ ao redor de 
$ t $, obtemos uma aproxima√ß√£o para $ \boldsymbol y(t+h) $ a partir da informa√ß√£o dos
valores de $ \boldsymbol y $ e suas derivadas de ordens at√© $ m $ em $ t $:
\begin{equation*}\label{E:Taylor}
\boldsymbol y(t+h) \approx \boldsymbol y(t) + h\,\boldsymbol y'(t) + \frac{h^2}{2}\,\boldsymbol y''(t) + \cdots + \frac{h^m}{m!}\boldsymbol y^{(m)}(t)\,. \tag{2}
\end{equation*}
Pela f√≥rmula para o resto no teorema de Taylor, o erro envolvido √© aproximadamente proporcional a $ \abs{h}^{m+1} $. Portanto, para $ \abs{h} $ pr√≥ximo de $ 0 $, quanto maior o valor de $ m $, melhor √© esta aproxima√ß√£o.

A estimativa dada por \eqref{E:Taylor} √© a base do *m√©todo de Taylor de ordem* $ m $, do qual o m√©todo de Euler √© o caso especial em que $ m = 1 $. Antes de detalh√°-lo, precisamos expressar as sucessivas derivadas de $ \boldsymbol y $ em $ t $ usando apenas $ t $ e o valor de $ \boldsymbol y $ em $ t $.

## $ \S 2 $ O m√©todo de Taylor de ordem $ m $

### $ 2.1 $ Expressando $ \boldsymbol y^{(j)}(t) $ em termos de $ t $ e $ \boldsymbol y(t) $

Se $ \boldsymbol y $ √© a solu√ß√£o anal√≠tica do problema de valor inicial \eqref{E:ivp}, ent√£o
\begin{equation*}\label{E:derivative}
\boldsymbol y'(t) = \boldsymbol F\big(t, \boldsymbol y(t)\big)\,. \tag{3}
\end{equation*}
Assim, podemos determinar $ \boldsymbol y'(t) $ usando o valor de $ \boldsymbol y(t) $.

Derivando \eqref{E:derivative} com respeito a $ t $ usando a regra da cadeia, obtemos a seguinte f√≥rmula para $ \boldsymbol y'' $:
\begin{alignat*}{9}
\boldsymbol y''(t) &= \frac{d}{dt} \boldsymbol F\big(t,\boldsymbol y(t)\big) \\
&=\Bigg(\frac{\partial \boldsymbol F}{\partial t} \frac{dt}{dt} + \sum_{k=0}^{r-1} \frac{\partial \boldsymbol F}{\partial y_k} \frac{dy_k}{dt} \Bigg)\big(t,\boldsymbol y(t)\big) \\
&= \Bigg(\boldsymbol F_t + \sum_{k=0}^{r-1}\boldsymbol F_{y_k} f_k\Bigg)\big(t,\boldsymbol y(t)\big) \,.\label{E:T2} \tag{4}
\end{alignat*}
Aqui
$$
\boldsymbol F = (f_0,f_1,\dots,f_{r-1})\,
$$
e no √∫ltimo passo usamos \eqref{E:derivative} para substituir a $ k $-√©sima coordenada $ y_k' $ do lado esquerdo desta equa√ß√£o pela $ k $-√©sima coordenada $ f_k $ do lado direito. Observe que a express√£o \eqref{E:T2} n√£o envolve $ \boldsymbol y' $, apenas $ \boldsymbol y $.

Da mesma maneira, derivando sucessivamente \eqref{E:derivative} com respeito a $ t $, obtemos uma f√≥rmula para $ \boldsymbol y^{(j)} $ da forma:
\begin{equation*}\label{E:T}
\boldsymbol y^{(j)}(t) = \boldsymbol T_j\big(t,\boldsymbol y(t)\big) \qquad (j = 1, 2, \dots)\,. \tag{5}
\end{equation*}
A express√£o √† direita √© complicada, pois envolve todas as derivadas parciais de $ \boldsymbol F $ de ordem $ \le j - 1 $. Mas o ponto principal √© que, por indu√ß√£o, podemos escrev√™-la de modo que ela n√£o envolva 
$$
\boldsymbol y'(t),\ \boldsymbol y''(t), \cdots,\ \boldsymbol y^{(j-1)}(t)\,,
$$
apenas $ t $ e $ \boldsymbol y(t) $, como j√° indicado pela nota√ß√£o.

### $ 2.2 $ Descri√ß√£o do m√©todo de Taylor de ordem $ m $

Sejam $ N \ge 1 $ um inteiro, $ h \ne 0 $ o **tamanho do passo** e 
$$
t_i = t_0 + ih \qquad (i = 0, 1, \dots, N)
$$
os **nodos** correspondentes.

No **m√©todo de Taylor de ordem $ m $**, definimos a aproxima√ß√£o $ \boldsymbol y_i $ para o valor preciso $ \boldsymbol y(t_i) $ da solu√ß√£o anal√≠tica de \eqref{E:ivp} usando a estimativa $ \boldsymbol y_{i-1} $ no nodo anterior e a aproxima√ß√£o polinomial de ordem $ m $ em \eqref{E:Taylor}, mas substituindo $ \boldsymbol y^{(j)} $ pelo lado direito em \eqref{E:T} para cada $ j $:
\begin{equation*}\label{E:m}
\boxed{\boldsymbol y_{i} = \boldsymbol y_{i-1} + h\,\boldsymbol T_1\big(t_{i-1},\boldsymbol y_{i-1}\big) +
\frac{h^2}{2}\,\boldsymbol T_2\big(t_{i-1},\boldsymbol y_{i-1}\big) + \cdots + \frac{h^m}{m!}\boldsymbol T_m\big(t_{i-1},\boldsymbol y_{i-1}\big)}\tag{6}
\end{equation*}
onde $t_i = t_0 + ih$ para cada $ i = 1, \dots, N $.

üìù Tomando $ m = 1 $, recuperamos o m√©todo de Euler.

üìù A f√≥rmula geral acima ainda n√£o √© √∫til porque os $ \boldsymbol T_j $ n√£o foram determinados. Faremos isto nas se√ß√µes seguintes em casos especiais.

## $ \S 3 $ F√≥rmula do m√©todo de Taylor de ordem $ 2 $ para um problema de valor inicial vetorial

Usando a nota√ß√£o da $ \S 2 $, conclu√≠mos de \eqref{E:derivative} e \eqref{E:T2} que
$$
\boxed{\boldsymbol T_1 = \boldsymbol F \quad \text{e} \quad \boldsymbol T_2 = \boldsymbol F_t + \sum_{k=0}^{r-1}\boldsymbol F_{y_k} f_k}
$$

Substituindo estas express√µes em \eqref{E:m}, obtemos finalmente a f√≥rmula para o **m√©todo de Taylor de ordem** $ 2 $:
\begin{equation*}
\boxed{\boldsymbol y_{i} = \boldsymbol y_{i-1} + h\,\boldsymbol F\big(t_{i-1},\boldsymbol y_{i-1}\big) +
\frac{h^2}{2}\,\Bigg( \boldsymbol F_t + \sum_{k=0}^{r-1}\boldsymbol F_{y_k} f_k \Bigg)\big(t_{i-1},\boldsymbol y_{i-1}\big)}% \qquad \text{onde }\boldsymbol F = (f_0,f_1,\dots,f_{r-1})
\end{equation*}

## $ \S 4 $ F√≥rmulas do m√©todo de Taylor de ordens $ 2 $  e $ 3 $ para um problema de valor inicial escalar

Para simplificar a situa√ß√£o, considere um PVI de primeira ordem *escalar* (ou seja, com $ r = 1 $):
\begin{equation*}
\begin{cases}
y' = f(t, y) \\
y(t_0) = y_0
\end{cases}
\end{equation*}
Como caso especial das f√≥rmulas deduzidas na $ \S 3 $, sabemos que:
$$
\boxed{T_1 = f}
$$
e
$$
\boxed{\displaystyle{T_2 = f_t + f_y\,f}}
$$

Derivando $ T_2\big(t,y(t)\big) $ com respeito a $ t $ com ajuda da regra da cadeia (e omitindo os par√¢metros por concis√£o), deduzimos que:
\begin{alignat*}{9}
T_3 &=\big(f_t + f_y\,f\big)_t + \big(f_t + f_y\,f\big)_y\,y' \\
&=\big(f_t + f_y\,f\big)_t + \big(f_t + f_y\,f\big)_y\,f \\
&= \boxed{f_{tt} + f_t\,f_y + 2f_{ty}\,f + f_{yy}\,f^2 +   f_y^2\,f}
\end{alignat*}

Substituindo as express√µes para  $ T_1,\,T_2,\,T_3 $ em \eqref{E:m} obtemos as f√≥rmulas dos m√©todos de Taylor de ordens $ 1 $, $ 2 $ e $ 3 $ *no caso escalar*.

Continuando a derivar poder√≠amos encontrar $ T_j $ para $ j $ qualquer, mas n√£o o faremos pelo motivo j√° indicado no caso geral: as express√µes resultantes s√£o complicadas demais para serem valiosas.

## $ \S 5 $ Implementa√ß√£o do m√©todo de Taylor de ordem $ 2 $ no caso escalar

*Para aplicar o m√©todo de Taylor de ordem $ m $, precisamos conhecer todas as derivadas parciais de ordem at√© $ m - 1 $ da fun√ß√£o $ f $ (ou $ \boldsymbol F $) que define a equa√ß√£o diferencial*. Temos duas op√ß√µes na implementa√ß√£o:
* Podemos calcular estas derivadas analiticamente e pass√°-las como entradas;
* Podemos fornecer apenas a fun√ß√£o $ f $ como entrada, e deixar a determina√ß√£o das derivadas para o computador, usando algum m√©todo de diferencia√ß√£o num√©rica.

A segunda op√ß√£o √© muito mais conveniente e geral, j√° que n√£o precisamos calcular n√≥s mesmos estas derivadas para cada EDO √† qual o m√©todo ser√° aplicado. Por√©m, ela tem a desvantagem que, quando as derivadas n√£o s√£o determinadas simbolicamente, estamos introduzindo erros j√° na *f√≥rmula* que define as aproxima√ß√µes. Para piorar, os m√©todos mais comuns de diferencia√ß√£o num√©rica tem precis√£o relativamente baixa.

Por este motivo, na implementa√ß√£o abaixo do m√©todo de Taylor de ordem $ 2 $ no caso escalar, √© o usu√°rio quem deve calcular e fornecer as derivadas parciais $ f_t $ e $ f_y $.

```python
def taylor_2(f, f_t, f_y, t_0, y_0, T, h):
    """
    Fornece uma solu√ß√£o num√©rica de um PVI de 1a ordem da forma:
        y' = f(t, y), y(t_0) = y_0
    usando o m√©todo de Taylor de ordem 2.
    Entradas:
        * A fun√ß√£o escalar f = f(t, y) que define a EDO.
        * A derivada f_t de f com respeito a t.
        * A derivada f_y de f com respeito a y.
        * O instante inicial t_0.
        * O escalar y_0 tal que y(t_0) = y_0 √© a condi√ß√£o inicial.
        * O instante (nodo) T final.
        * O tamanho do passo h, que deve ter o mesmo sinal que T - t_0.
    Sa√≠das:
        * Um array ts = [t_0, t_1, ..., t_N = T] onde t_i = t_0 + ih
          para i = 0, 1, ..., N - 1 e T - t_{N - 1} <= h.
        * Um array correspondente ys = [y_0, y_1, ..., y_N] contendo
          as aproxima√ß√µes para a solu√ß√£o anal√≠tica em t_i.
    """
    import numpy as np
    
    def T_1(t, y):
        return f(t, y)
    
    def T_2(t, y):
        return f_t(t, y) + f_y(t, y) * f(t, y)
        
    sinal = np.sign(h)
    assert np.sign(T - t_0) == sinal
    ts = [t_0]
    ys = [y_0]
    t = t_0
    y = y_0
    
    # Enquanto n√£o atingimos T, atualizamos Y = Y_i usando a f√≥rmula
    # do m√©todo. Depois incrementamos (ou decrementamos) t por h.
    while sinal * (T - t) > 0:    # t < T (se h > 0) ou T < t (se h < 0)
        # Se falta menos que |h| para chegar em T, usamos um passo menor:
        h = sinal * min(abs(h), abs(T - t))
        y += h * (T_1(t, y) + h * T_2(t, y) / 2)
        t += h
        ys.append(y)
        ts.append(t)
    
    return np.array(ts), np.array(ys)
```

## $ \S 6 $ Estimativa para o erro no m√©todo de Taylor de ordem $ m $

Pelo teorema de Taylor com resto, o **erro local**, i.e., o erro envolvido no *primeiro* passo, √© da ordem de $ \abs{h}^{m+1} $:
$$
\boxed{\boldsymbol E_1 = \boldsymbol y(t_1) - \boldsymbol y_1 \in O(h^{m+1})}
$$
Assim como para o m√©todo de Euler, uma estimativa do **erro global**, ou seja, da diferen√ßa
$ \boldsymbol E_N = \boldsymbol y(t_N) - \boldsymbol y_N $
entre o valor exato e o valor aproximado no *√∫ltimo* nodo $ t_N $, pode ser obtida assumindo-se que o erro a cada passo √© constante. Ap√≥s $ N $ passos, √© de se esperar que o erro global seja da ordem de $ \abs{h}^m $:
$$
\boldsymbol E_N \approx N\, \boldsymbol E_1 = \frac{t_N - t_0}{h}\,\boldsymbol E_1.
$$
Em s√≠mbolos:
$$
\boxed{\boldsymbol E_N = \boldsymbol y(t_N) - \boldsymbol y_N \in O(h^m)}
$$

üìù Segue imediatamente que para $ \abs{h} $ pequeno, quanto maior o valor de $ m $, mais precisa ser√° a solu√ß√£o num√©rica fornecida pelo m√©todo de Taylor. Por√©m esta maior precis√£o vem ao custo da determina√ß√£o (anal√≠tica ou num√©rica) da express√£o dos $ \boldsymbol T_j $ que aparecem na f√≥rmula geral para cada $ j $ de $ 1 $ at√© $ m $. A dificuldade desta determina√ß√£o cresce rapidamente conforme $ m $ aumenta. Por isto na pr√°tica outros m√©todos para resolu√ß√£o de problemas de valor inicial s√£o preferidos.

## $ \S 8 $ Procedimentos para impress√£o e plotagem da solu√ß√£o num√©rica

```python
def imprime_solucao(ts, Ys, freq):
    """
    Dados dois arrays ts e Ys de mesmo comprimento, imprime um
    a cada 'freq' de seus valores por linha, na forma de uma
    tabela. Os 0-√©simos e √∫ltimos valores sempre s√£o impressos.
    """
    def imprime_cabecalho(r):
        """
        Imprime o cabe√ßalho da tabela.
        """
        print("\n       t              ", end="")
        for j in range(r):
            print(f"y_{j}            ", end="")
        print()
        for j in range(r + 1):
            print("---------------", end="")
        print()
        
        
    def imprime_linha(t, Y, r):
        """
        Imprime uma das linhas da tabela.
        """
        print("{:13.5f}".format(t), end="")
        # Para dimens√£o r > 1:
        if r > 1:
            for j in range(r):
                print("  {:13.5f}".format(Y[j]), end="")
            print()
        # Para dimens√£o r = 1:
        else:
            print("  {:13.5f}".format(Y))
    
    N = len(Ys) - 1        # N + 1 √© o n√∫mero de nodos; N o de passos.
    try:
        r = len(Ys[0])     # Dimens√£o de cada entrada Y de Ys.
    except TypeError:
        r = 1
    if freq == 0:          # Se freq == 0, imprime apenas os √∫ltimos valores.
        freq = N
    imprime_cabecalho(r)
    for i in range(0, N + 1, freq):
        imprime_linha(ts[i], Ys[i], r)
    if i != N:
        imprime_linha(ts[N], Ys[N], r)
    for j in range(r + 1):
            print("---------------", end="")
    print()
        
    return None
```

```python
def compara_solucao(f, t_0, y_0, ts, ys):
    """ Representa graficamente as seguintes solu√ß√µes do PVI (escalar)
            y' = f(t, y), y(t_0) = y_0
    (1) Uma solu√ß√£o num√©rica dada pelos dois arrays (ou listas)
        ts e ys, calculada previamente por um m√©todo qualquer.
    (2) A solu√ß√£o obtida pela rotina `odeint` da biblioteca SciPy.
    Entradas:
        * A fun√ß√£o escalar f(t, y) que define a EDO.
        * O instante inicial t_0.
        * O escalar y_0 tal que y(t_0) = y_0 √© a condi√ß√£o inicial.
        * A lista ts dos nodos t_i.
        * A lista ys das aproxima√ß√µes y_i obtidas.
    """
    import matplotlib.pyplot as plt
    import numpy as np
    from scipy.integrate import odeint
    
    
    ts_exata = np.linspace(t_0, T, num=201)
    ys_exata = odeint(f, y_0, ts_exata, tfirst=True)
    
    plt.plot(ts, ys, 'o')
    plt.plot(ts_exata, ys_exata, '-')
    plt.grid(True)
    plt.xlabel('t')
    plt.ylabel('y')
    plt.legend(("Num√©rica", "Exata"), loc=0)
    plt.show()
    
    return None
```

# M√©todos de Taylor

## $ \S 1 $ Recorda√ß√£o do m√©todo de Euler

Pelo menos para $ h $ pequeno, uma fun√ß√£o
$$
\mbf y (t) = \big(y_0(t),\, y_1(t), \cdots,\, y_{r-1}(t)\big)
$$
√© bem aproximada pelo seu polin√¥mio de Taylor de ordem 1:
\begin{equation*}
\mbf y(t+h) \approx \mbf y(t) + h\,\mbf y'(t).
\end{equation*}

Portanto, se $ \mbf y $ √© a solu√ß√£o de um problema de valor inicial de primeira ordem em $ \mathbb R^r $:
\begin{equation*}\label{E:ivp}
\begin{cases}
\mbf y' = \mbf F(t, \mbf y) \\
\mbf y(t_0) = \mbf y_0
\end{cases}
\quad \text{onde} \quad \mbf y = \big(y_0, y_1, \dots, y_{r-1}\big)\,, \tag{1}
\end{equation*}
ent√£o
\begin{equation*}
\mbf y(t+h) \approx \mbf y(t) + h\,\mbf F\big(t, \mbf y(t)\big).
\end{equation*}

Esta rela√ß√£o √© a base do *m√©todo de Euler* do caderno anterior. Partindo de $ \mbf y(t_0) = \mbf y_0 $ e procedendo em incrementos de tamanho $ h $, conseguimos uma estimativa $ \mbf y_i $ para o valor de $ \mbf y $ no nodo $ t_{i} $ a partir da estimativa $ \mbf y_{i - 1} $ no nodo anterior:
\begin{equation*}
\mbf y_{i} = \mbf y_{i-1} + h\,\mbf F(t_{i-1}, \mbf y_{i-1}) \quad \text{onde $t_i = t_0 + ih$} \quad (i = 1, \dots, N).
\end{equation*}

## $ \S 2 $ Id√©ia do m√©todo de Taylor de ordem $ m $

Usando o polin√¥mio de Taylor de ordem $ m \ge 1 $, obtemos uma aproxima√ß√£o ainda mais precisa para $ \mbf y(t+h) $ a partir da informa√ß√£o sobre $ \mbf y $ e suas derivadas em $ t $:
\begin{equation*}\label{E:Taylor}
\mbf y(t+h) \approx \mbf y(t) + h\,\mbf y'(t) + \frac{h^2}{2}\,\mbf y'(t) + \cdots + \frac{h^m}{m!}\mbf y^{(m)}(t)\,. \tag{3}
\end{equation*}
Pela f√≥rmula para o resto no teorema de Taylor, o erro envolvido √© aproximadamente proporcional a $ h^{m+1} $.

Se $ \mbf y $ √© a solu√ß√£o anal√≠tica do problema de valor inicial \eqref{E:ivp}, ent√£o
\begin{equation*}\label{E:derivative}
\mbf y'(t) = \mbf F\big(t, \mbf y(t)\big). \tag{4}
\end{equation*}
Derivando com respeito a $ t $ usando a regra da cadeia, obtemos uma f√≥rmula para $ \mbf y'' $:
$$
\mbf y''(t) = \mbf T_2\big(t,\mbf y(t)\big)\,.
$$
Os detalhes podem ser encontrados na $ \S 5 $, mas o que importa aqui √© que esta express√£o envolve apenas $ \mbf F $, suas derivadas parciais de primeira ordem, $ t $ e $ \mbf y(t) $, mas n√£o $ \mbf y'(t) $, j√° que por \eqref{E:derivative} esta √∫ltima pode ser expressa como $ \mbf F\big(t,y(t)\big) $.

Da mesma maneira, derivando sucessivamente \eqref{E:derivative} com respeito a $ t $, obtemos uma express√£o para $ \mbf y^{(j)} $ da forma:
\begin{equation*}\label{E:T}
\mbf y^{(j)}(t) = \mbf T_j\big(t,\mbf y(t)\big) \qquad (j = 1, 2, \dots)\,. \tag{5}
\end{equation*}
A express√£o √† direita √© complicada, pois envolve todas as derivadas parciais de $ \mbf F $ de ordem $ \le j - 1 $. Contudo, por indu√ß√£o, podemos escrev√™-la de modo que *ela n√£o dependa de* 
$$
\mbf y'(t),\ \mbf y''(t), \cdots,\ \mbf y^{(j-1)}(t),
$$
apenas de $ \mbf y(t) $, como j√° indicado pela nota√ß√£o.

Sejam $ N \ge 1 $ um inteiro, $ h > 0 $ o **tamanho do passo** e 
$$
t_i = t_0 + ih \qquad (i = 0, 1, \dots, N)
$$
os **nodos** correspondentes. No **m√©todo de Taylor de ordem $ m $**, definimos a aproxima√ß√£o $ \mbf y_i $ para o valor preciso $ \mbf y(t_i) $ da solu√ß√£o de \eqref{E:ivp} usando a aproxima√ß√£o polinomial de ordem $ m $ em \eqref{E:Taylor}, mas substituindo $ \mbf y^{(j)} $ pelo lado direito em \eqref{E:T} para cada $ j $:
\begin{equation*}\label{E:method}
\boxed{\mbf y_{i} = \mbf y_{i-1} + h\,\mbf T_1\big(t_{i-1},\mbf y_{i-1}\big) +
\frac{h^2}{2}\,\mbf T_2\big(t_{i-1},\mbf y_{i-1}\big) + \cdots + \frac{h^m}{m!}\mbf T_m\big(t_{i-1},\mbf y_{i-1}\big)} \tag{6}
\end{equation*}
onde $t_i = t_0 + ih$ para cada $ i = 1, \dots, N $.

üìù Tomando $ m = 1 $, recuperamos o m√©todo de Euler.

üìù A f√≥rmula acima n√£o √© muito √∫til porque os $ \mbf T_j $ n√£o foram determinados. Faremos isto nas se√ß√µes seguintes em casos especiais.

## $ \S 3 $ F√≥rmulas do m√©todo de Taylor de ordem $ m = 2 $  e $ m = 3 $ para um problema de valor inicial escalar

Para simplificar, considere inicialmente um PVI de primeira ordem *escalar* (ou seja, com $ r = 1 $):
\begin{equation*}
\begin{cases}
y' = f(t, y) \\
y(t_0) = y_0
\end{cases}
\end{equation*}
Derivando a rela√ß√£o 
$$
y'(t) = f(t, y(t)) =: T_1\big(t,y(t)\big)
$$
com respeito a $ t $ usando a regra da cadeia, deduzimos que:
\begin{alignat*}{9}
y''(t) &= f_t\big(t, y(t)\big) + f_y\big(t, y(t)\big)\,y'(t) \\
& = f_t\big(t, y(t)\big) + f_y\big(t, y(t)\big)\,f\big(t, y(t)\big).
\end{alignat*}
Ou, omitindo os par√¢metros:
\begin{equation*}
y'' = f_t + f_y\,f
\end{equation*}
Na nota√ß√£o da se√ß√£o anterior:
\begin{equation*}\label{E:T1}
\boxed{T_2 = f_t + f_y\,f.}
\end{equation*}
Derivando mais uma vez com ajuda da regra da cadeia, obtemos:
\begin{alignat*}{9}
y''' = y^{(3)} &= f_{tt} + 2f_{ty}\,f + f_{yy}\,f^2 + \\
&+ f_t\,f_y +  f_y^2\,f.
\end{alignat*}
Assim:
\begin{equation*}\label{E:T2}
\boxed{T_3 = f_{tt} + 2f_{ty}\,f + f_{yy}\,f^2 +
+ f_t\,f_y +  f_y^2\,f.}
\end{equation*}

Continuando, poder√≠amos encontrar $ T_j $ para $ j $ qualquer, mas n√£o o faremos.

üìù Observe que estas f√≥rmulas s√£o v√°lidas apenas se $ y $ √© uma fun√ß√£o escalar ($ r = 1 $).

## $ \S 3 $ Implementa√ß√£o do m√©todo de Taylor de ordem $ 2 $ (caso escalar)

```python
def taylor_2(f, f_t, f_y, t_0, Y_0, T, h):
    """
    Fornece uma solu√ß√£o num√©rica de um problema de valor inicial de
    1a ordem da forma y' = f(t, y), y(t_0) = y_0 usando o m√©todo de Taylor
    de ordem 2.
    Entradas:
        * A fun√ß√£o escalar f(t, y) que define a EDO.
        * A derivada f_t de f com respeito a t.
        * A derivada f_y de f com respeito a y.
        * O instante inicial t_0.
        * Um vetor ou escalar y_0 tal que y(t_0) = y_0 √© a condi√ß√£o inicial.
        * O instante T final.
        * O tamanho do passo h.
    Sa√≠das:
        * Um array ts = [t_0, t_1, ..., t_N = T] onde t_i = t_0 + ih
          para i = 0, 1, ..., N - 1 e T - t_{N - 1} <= h.
        * Um array correspondente ys = [y_0, y_1, ..., y_N] contendo
          as aproxima√ß√µes para a solu√ß√£o anal√≠tica em t_i.
    """
    import numpy as np
    
    def T_1(t, y):
        return f(t, y)
    
    def T_2(t, y):
        return f_t(t, y) + f_y(t, y) * f(t, y)
    
    ts = [t_0]
    ys = [y_0]
    t = t_0
    y = y_0
    while t < T:
        h = min(h, T - t)
        y += h * (T_1(t, y) + h * T_2(t, y) / 2)
        t += h
        ys.append(y)
        ts.append(t)
    
    return np.array(ts), np.array(ys)
```

## $ \S 4 $ F√≥rmula do m√©todo de Taylor de ordem $ m = 2 $ para um problema de valor inicial vetorial

Suponha novamente que $ \mbf y $ seja a solu√ß√£o anal√≠tica da equa√ß√£o diferencial vetorial
$$
\mbf y' = \mbf F(t, \mbf y)=: \mbf T_1(t, y).
$$
Derivando esta igualdade com respeito a $ t $ usando a regra da cadeia, obtemos uma f√≥rmula para $ \mbf y'' $ que envolve $ \mbf F $, suas derivadas parciais de primeira ordem, e $ \mbf y $, mas n√£o $ \mbf y' $:
\begin{alignat*}{9}
\mbf y'' &= \mbf F_t + \sum_{k=0}^{r-1}\mbf F_{y_k} y_k' \\
&= \mbf F_t + \sum_{k=0}^{r-1}\mbf F_{y_k} f_k \qquad \text{onde }\mbf F = (f_0,f_1,\dots,f_{r-1}).
\end{alignat*}
Aqui $ \mbf F_t $ denota a derivada parcial de $ \mbf F $ com respeito a $ t $ e $ \mbf F_{y_k} $ aquela com respeito √† $ k $-√©sima coordenada de $ \mbf y $.
Na nota√ß√£o da $ \S 2 $, conclu√≠mos portanto que
$$
\boxed{\mbf T_2 = \mbf F_t + \sum_{k=0}^{r-1}\mbf F_{k_j} f_k} \qquad \big(\mbf F = (f_0,f_1,\dots,f_{r-1})\big)\,.
$$
Substituindo as express√µes para $ \mbf T_1 $ e $ \mbf T_ 2$ na eq. (6) da $ \S 2 $, obtemos a f√≥rmula para o m√©todo de Taylor de ordem $ 2 $ no caso vetorial.

üìù Derivando sucessivamente a express√£o para $ \mbf y'' $ poder√≠amos deduzir as f√≥rmulas para qualquer $ \mbf T_j $, mas, novamente, n√£o o faremos aqui.

## $ \S 5 $ Estimativa para o erro no m√©todo de Taylor de ordem $ m $

Pelo teorema de Taylor com resto, o **erro local**, i.e., o erro envolvido no primeiro passo, √© da ordem de $ h^{m+1} $:
$$
\boxed{\mbf E_1 = \mbf y(t_1) - \mbf y_1 \in O(h^{m+1})}
$$
Assim como para o m√©todo de Euler, uma estimativa do **erro global**, ou seja, da diferen√ßa
$ \mbf E_N = \mbf y(t_N) - \mbf y_N $
entre o valor exato e o valor aproximado no √∫ltimo nodo $ t_N $, pode ser obtida assumindo-se que o erro a cada passo √© constante. Ap√≥s $ N $ passos, √© de se esperar que o erro global seja da ordem de $ h^m $:
$$
\mbf E_N \approx N \mbf E_1 = \frac{t_N - t_0}{h}\,\mbf E_1.
$$
Em s√≠mbolos:
$$
\boxed{\mbf E_N = \mbf y(t_N) - \mbf y_N \in O(h^m)}
$$

üìù Segue imediatamente que quanto maior o valor de $ m $, mais precisa ser√° a solu√ß√£o num√©rica fornecida pelo m√©todo de Taylor. Por√©m esta maior precis√£o vem ao custo da determina√ß√£o da express√£o expl√≠cita para os $ \mbf T_j $ para cada $ j $ de $ 1 $ at√© $ m $, que aparecem na f√≥rmula. A dificuldade desta determina√ß√£o cresce rapidamente conforme $ m $ aumenta, pois requer o c√°lculo de todas as derivadas parciais de $ \mbf F $ de ordem at√© $ m - 1 $.

## $ \S 7 $ Procedimentos para impress√£o e plotagem da solu√ß√£o num√©rica

```python
def imprime_solucao(ts, Ys, freq):
    """
    Dados dois arrays ts e Ys de mesmo comprimento, imprime um
    a cada 'freq' de seus valores por linha, na forma de uma
    tabela. Os 0-√©simos e √∫ltimos valores sempre s√£o impressos.
    """
    def imprime_cabecalho(r):
        """
        Imprime o cabe√ßalho da tabela.
        """
        print("\n        t            ", end="")
        for j in range(r):
            print(f"y_{j}          ", end="")
        print()
        for j in range(r + 1):
            print("---------------", end="")
        print()
        
        
    def imprime_linha(t, Y, r):
        """
        Imprime uma das linhas da tabela.
        """
        print("{:13.4f}".format(t), end="")
        # Para dimens√£o r > 1:
        if r > 1:
            for j in range(r):
                print("{:13.4f}".format(Y[j]), end="")
            print()
        # Para dimens√£o r = 1:
        else:
            print("{:13.4f}".format(Y))
    
    
    N = len(Ys) - 1        # N + 1 √© o n√∫mero de nodos; N o de passos.
    try:
        r = len(Ys[0])     # Dimens√£o de cada entrada Y de Ys.
    except TypeError:
        r = 1
    if freq == 0:          # Se freq == 0, imprime apenas os √∫ltimos valores.
        freq = N
    imprime_cabecalho(r)
    for i in range(0, N + 1, freq):
        imprime_linha(ts[i], Ys[i], r)
    if i != N:
        imprime_linha(ts[N], Ys[N], r)
        
    return None
```

```python
def plota_poligonal(ts, ys):
    """
    Dados arrays ou listas num√©ricos ts e ys de mesmo comprimento,
    plota a curva poligonal de v√©rtices (ts[i], ys[i]).
    """
    import matplotlib.pyplot as plt
    
    plt.plot(ts, ys, '-')
    plt.grid(True)
    plt.xlabel('t')
    plt.ylabel('y')
    plt.show()
    
    return None
```

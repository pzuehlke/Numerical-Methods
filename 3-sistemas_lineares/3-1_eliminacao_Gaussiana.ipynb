```python
import numpy as np


def doolittle_factorization(A: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    """
    Performs the Doolittle factorization of a matrix, decomposing it into a 
    product of a lower triangular and an upper triangular matrix.
    Parameters:
        A (np.ndarray): The input square matrix to be factorized.
    Returns:
        L, U: The lower and upper triangular matrices, L and U, 
              respectively, such that A = LU.
    Raises:
        ValueError: If a zero pivot is encountered during the row reduction, 
                    indicating that the input matrix cannot be factorized using 
                    this method.
    """
    n = np.shape(A)[0]   # Extract the length of the columns/rows of A.
    L = np.eye(n)        # n x n identity matrix.
    for j in range(0, n - 1):
        if A[j][j] == 0:
            raise ValueError(f"Encountered a zero pivot in the {j}-th step."
                               "The given matrix cannot be row-reduced!")
        for i in range(j + 1, n):
            factor = A[i][j] / A[j][j]
            L[i][j] = factor
            A[i, :] = A[i, :] - factor * A[j, :]
    U = A.copy()
    return L, U
```

```python
def determinant_row_reduction(A: np.ndarray) -> float:
    """
    Computes the determinant of a matrix using row reduction.
    Parameters:
        * A: The input square matrix whose determinant is to be computed.
    Returns:
        * The determinant of the matrix.
    """
    _, U = doolittle_factorization(A)
    return np.prod(np.diag(U))
```

```python
def inverse_row_reduction(A: np.ndarray) -> np.ndarray:
    """
    Computes the inverse of a matrix using row reduction.
    Parameters:
        * A: The input square matrix to be inverted.
    Returns:
        * The inverse of the input matrix.
    """
    n = np.shape(A)[0]
    AI = np.hstack([A, np.eye(n)])
    for i in range(n):
        factor = AI[i][i]
        AI[i, :] = AI[i, :] / factor
        for j in range(n):
            if i != j:
                factor = AI[j][i]
                AI[j, :] = AI[j, :] - factor * AI[i, :]
    return AI[:, n:]
```

```python
def solve_linear_systems(A: np.ndarray, B: np.ndarray) -> np.ndarray:
    """
    Solves multiple linear systems of the form Ax = B, where each column of B 
    defines a system to be solved (for fixed A). If B is a single column-vector, 
    then it solves a single linear system.
    Parameters:
        * A: The input square matrix defining the linear systems.
        * B: The matrix of right-hand sides. Each column defines a different system.
    Returns:
        * The solution matrix. Each column is the solution to the corresponding system.
    """
    n = np.shape(A)[0]
    AB = np.hstack([A, B])
    for i in range(n):
        factor = AB[i][i]
        AB[i, :] = AB[i, :] / factor
        for j in range(n):
            if i != j:
                factor = AB[j][i]
                AB[j, :] = AB[j, :] - factor * AB[i, :]
    return AB[:, n:]
```

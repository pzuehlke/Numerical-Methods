## $ \S 1 $ Matrizes elementares

Uma outra raz√£o para usar a forma matricial do escalonamento √© a seguinte:
_Aplicar uma opera√ß√£o elementar a um sistema $ \mathbf{A} \mathbf x =
\mathbf{b} $ corresponde a multiplicar a matriz aumentada $ [\mathbf{A}|
\mathbf{b}] $ √† esquerda por uma matriz apropriada._

__Defini√ß√£o:__ A multiplica√ß√£o √† esquerda pelas seguintes matrizes quadradas
$ n \times n $, chamadas de __matrizes elementares__, correspondem respectivamente √†s
opera√ß√µes elementares $ (S) $, $ (T) $ e $ (M) $:

+  $ (S) $ A opera√ß√£o $ L_{i} \leftarrow L_i - \lambda_{ij} L_j $, para $ i > j $, corresponde √† matriz
   $ \mathbf E_{ij}(-\lambda_{ij}) $ que possui todas as entradas nulas exceto pelas da diagonal,
   que s√£o iguais a $ 1 $, e a da posi√ß√£o $ (i,j) $, que √© igual a $ -\lambda_{ij} $.
+  $ (T) $ A opera√ß√£o de troca $ L_i \leftrightarrow L_j $ corresponde √†
   __matriz-permuta√ß√£o__ $ \mathbf P_{ij} $ obtida a partir de $ I_n $ intercambiando-se
   as linhas $ i $ e $ j $.
+  $ (M) $ A opera√ß√£o $ L_i \leftarrow \mu L_i $
   $ (\mu \ne 0) $ corresponde √† matriz diagonal obtida a partir de $ I_n $ (a matriz
   identidade $ n \times n $) fazendo a entrada $ (i,i) $ igual a $ \mu $.

__Exemplo 2:__ As matrizes
$$
	\mathbf E_{21}(-\lambda)=\begin{bmatrix}
	 1 & 0 & 0 \\
	 -\lambda & 1 & 0 \\
	 0 & 0 & 1
	\end{bmatrix},
	\quad
	\mathbf P_{13}=\begin{bmatrix}
	0 & 0 &1 \\
	0 & 1 & 0 \\
	1 & 0 & 0
	\end{bmatrix}
	\quad \text{e} \quad
	\begin{bmatrix}
	 1 & 0 & 0\\
	 0 & \mu & 0\\
	 0 & 0 & 1
	\end{bmatrix}
$$
s√£o elementares de tipos $(S) $, $ (T) $ e $ (M) $, respectivamente. De fato, se 
$$
	\mathbf A=\begin{bmatrix}
		a_1 & a_2 & a_3 \\
		b_1 & b_2 & b_2 \\
		c_1 & c_2 & c_3 
	\end{bmatrix}
$$
√© uma matriz $3\times 3$, ent√£o
$$
	\mathbf E_{21}(-\lambda)\,\mathbf A=\begin{bmatrix}
		a_1 & a_2 & a_3  \\
		b_1-\lambda a_1 & b_2 -\lambda a_2 & b_3-\lambda a_3 \\
		c_1 & c_2 & c_3  
	\end{bmatrix}
	\quad \text{e} \quad \mathbf P_{13}\, \mathbf A =
	\begin{bmatrix}
		c_1 & c_2 & c_3 \\
		b_1 & b_2 & b_2 \\
		a_1 & a_2 & a_3
	\end{bmatrix}\,,
$$
e similarmente para a terceira matriz elementar. Como n√£o √© necess√°rio utilizar
qualquer opera√ß√£o do tipo $ (M) $ para se escalonar um sistema, a partir de agora
vamos considerar apenas matrizes elementares de tipos $ (S) $ e $ (T) $.

Observe que todas as matrizes elementares s√£o invert√≠veis: Uma matriz de
permuta√ß√£o de linhas √© a inversa dela mesma e a matriz inversa de $E_{ij}(-\lambda)$
√© $E_{ij}(\lambda)$. Por exemplo:
$$
	\begin{bmatrix}
	 1 & 0 & 0 \\
	 0 & 1 & 0 \\
	0 & -7 & 1 
	\end{bmatrix}	\begin{bmatrix}
	 1 & 0 & 0 \\
	 0 & 1 & 0 \\
	0 & +7 & 1 
	\end{bmatrix}=\begin{bmatrix}
	 1 &0 & 0 \\
	 0 & 1 & 0 \\
	 0 & 0 & 1 
	\end{bmatrix},
$$
ou seja, $E_{32}(+7)E_{32}(-7)=I$.

Uma matriz quadrada √© __triangular inferior__ se todas as suas entradas acima da
diagonal s√£o nulas. A defini√ß√£o de matriz __triangular superior__ √© an√°loga.
√â f√°cil verificar que a soma e o produto de matrizes triangules inferiores (resp. superiores)
√© do mesmo tipo. Al√©m disto, a inversa de uma matriz triangular inferior (superior) √© do mesmo
tipo, desde que esta inversa exista. O determinante de uma matriz triangular √© o produto das
suas entradas diagonais. Observe que as matrizes elementares de tipo $ (S) $ s√£o
triangulares inferiores pela exig√™ncia que $ j < i $.

üìù Matriz triangulares superiores geralmente s√£o denotadas por $ \mathbf U $ (de
"_upper triangular_"), e matriz triangulares inferiores por $ \mathbf L $ (de
"_lower triangular_").

## $ \S 2 $ Fatora√ß√£o LU

__Teorema 2.1:__ _Seja $ \mathbf A $ uma matriz quadrada n√£o-singular
(i.e., invert√≠vel) $ n\times n $ e suponha que seja poss√≠vel escalonar $ \mathbf A $ 
utilizando somente opera√ß√µes elementares do tipo $ 1 $. Ent√£o podemos
escrever $ \mathbf A = \mathbf L \mathbf U $, onde $ \mathbf L $ √© triangular
inferior com elementos diagonais iguais a $ 1 $ e $ \mathbf U $ √© triangular
superior com elementos diagonais n√£o-nulos. Esta fatora√ß√£o √© √∫nica. Al√©m disto:_
* _A matriz $ \mathbf U $ √© a matriz resultante do escalonamento de_ $ \mathbf A $.
* *A entrada na posi√ß√£o $ (i, j) $ de $ \mathbf L $ (com $ i > j $) √©
  exatamente o coeficiente $ \lambda_{ij} $ da opera√ß√£o elementar
  $ L_i \leftarrow L_i - \lambda_{ij} L_j $ que envolve as linhas $ i $ e $ j $
  no escalonamento*.

Esta fatora√ß√£o √© importante por v√°rios motivos, dentre eles o seguinte:
Se sabemos fatorar $\mathbf A $ como $ \mathbf A = \mathbf L \mathbf U $,
ent√£o podemos resolver imediatamente o sistema
$$\mathbf A \mathbf x= \mathbf L \mathbf U \mathbf x = \mathbf b \,. $$
De fato, neste caso a resolu√ß√£o do sistema se reduz √† do par de sistemas:
$$
	\begin{cases}
		\mathbf L\mathbf y= \mathbf b  \\
		\mathbf U \mathbf x= \mathbf y
	\end{cases}\,.
$$
Como $ \mathbf L $ √© triangular inferior, o primeiro sistema pode ser resolvido
imediatamente por retro-substitui√ß√£o. Analogamente, uma vez encontrado $ \mathbf
y $, como $ \mathbf U $ √© triangular superior, o segundo sistema pode ser
resolvido por substitui√ß√£o progressiva (direta).

Como o determinante de uma matriz triangular √© simplesmente o produto de suas
entradas diagonais, se $ \mathbf A = \mathbf L \mathbf U $ √© a decomposi√ß√£o
de Doolittle de $ \mathbf A $, ent√£o 
$$
\det(\mathbf A) = \det(\mathbf U)\,,
$$
j√° que as entradas diagonais de $ \mathbf L $ s√£o todas iguais a $ 1 $.

## $ \S 3 $ Fatora√ß√£o $ LU $ e c√°lculo do determinante


```python
def row_reduction(A):
    import numpy as np

    n = np.shape(A)[0]
    L = np.eye(n)
    for j in range(0, n - 1):
        if A[j][j] == 0:
            raise ValueError(f"Encountered a zero pivot in the {j}-th step."
                               "The given matrix cannot be row-reduced!")
        for i in range(j + 1, n):
            factor = A[i][j] / A[j][j]
            L[i][j] = factor
            A[i, :] = A[i, :] - factor * A[j, :]
    U = A.copy()
    return U, L
```

## $ \S 3 $ Normas de matrizes

Recorde que o conjunto das matrizes $ n \times n $ com entradas reais ou, equivalentemente, o conjunto de todas as transforma√ß√µes lineares $ \mathbb R^n \to \mathbb R^n $, √© um espa√ßo vetorial de dimens√£o $ n^2 $. **Prova:** Seja $ \boldsymbol x = (x_1, x_2, \dots, x_n) $. Ent√£o

**Lema 3.2:** *Para quaisquer matrizes $ n \times n $ $ \boldsymbol A $ e $ \boldsymbol B $, vale:*

# Introdução à Matplotlib

## $ \S 1 $ Sobre a Matplotlib

A __Matplotlib__ é uma biblioteca de visualização de dados em Python que permite
criar uma ampla variedade de visualizações de alta qualidade de maneira simples
e flexível. Ela foi originalmente desenvolvida por John D. Hunter no
início dos anos 2000 e, desde então, tornou-se uma das ferramentas de
visualização de dados mais populares nas comunidades científica e de análise de
dados.

A Matplotlib facilita a criação de diversos tipos de figuras, como gráficos de
linha, de barras, de dispersão, de setores, de contorno, de densidade e de
superfícies 3D, além de histogramas e até mesmo animações. A biblioteca também
oferece suporte a anotações, rótulos, legendas, escalas personalizadas e mapas
de cores. Outra característica importante é a integração com outras bibliotecas
de visualização e análise de dados em Python, como Seaborn, pandas e
scikit-learn. Os gráficos gerados com a Matplotlib podem ser exportados em em
diversos formatos, como PNG, SVG, PDF e EPS.

Matplotlib é amplamente utilizada em ciência de dados, análise de dados,
aprendizado de máquina, pesquisa científica, engenharia, finanças, ou qualquer
outra área em que a visualização de dados desempenhe um papel fundamental para
entender e comunicar informações complexas.

## $ \S 2 $ Desenhando gráficos de funções de uma variável

### $ 2.1 $ Desenhando o gráfico de uma função real de uma variável

Suponha que tenhamos definido uma função real de uma variável `f` em Python, por exemplo,
a função
$$
f(x) = x^2 + 2x - 3\,.
$$

```python
f = lambda x: x**2 + 2 * x - 3
```

Para visualizar seu gráfico utilizando a Matplotlib, primeiro precisamos
importá-la; em conjunção com ela, utilizaremos também o NumPy:

```python
import matplotlib.pyplot as plt
import numpy as np
```

Agora precisamos criar um _array_ consistindo das coordenadas $ x $ e outro contendo
as coordenadas $ y = f(x) $ correspondentes que serão plotadas.  No caso da
lista de coordenadas $ x $, para automatizar esta tarefa, utilizaremos quase
sempre a função `numpy.linspace`, que gera um array de valores uniformemente
no intervalo desejado. Ela tem a assinatura:
`(<valor inicial>, <valor final>, <número de pontos>)`. 
Por exemplo, para criar $ 11 $ pontos igualmente espaçados no intervalo
$ [0, 1] $, utilizamos a instrução:

```python
# Recorde que importamos o NumPy sob o nome 'np':
xs = np.linspace(0, 1, 11)
print(xs)
```

No nosso caso, gostaríamos de gerar um array com $ 101 $ pontos 
no intervalo de $ -4 $ a $ 3 $:

```python
xs = np.linspace(-4, 3, 101)    # Gerando 101 pontos em [-4, 3].
ys = f(xs)    # Construindo o array [y = f(x)] para cada x em xs.
```

Finalmente, utilizamos a função `matplotlib.pyplot.plot` para desenhar o gráfico
de `f`:

```python
# Recorde que importamos matplotlib.pyplot sob o nome 'plt'.
plt.plot(xs, ys)
```

Para melhorar a apresentação, podemos adicionar rótulos aos eixos-$ x $ e $ y $,
bem como um título ao gráfico; e para facilitar ainda mais a visualização, podemos
desenhar um grid retangular além dos eixos. Finalmente, a função
`matplotlib.pyplot.show` exibir o gráfico:

```python
plt.plot(xs, ys)        # Cria um plot dos pontos a serem exibidos.
plt.xlabel("eixo-x")    # Especificando o rótulo de eixo-x.
plt.ylabel("eixo-y")    # Especificando o rótulo do eixo-y.
# Especificando um título para o gráfico:
plt.title("Gráfico da função $ y = f(x) = x^2 + 2x - 3 $")
plt.grid(True)          # Cria um grid retangular.
plt.show()              # Exibe o gráfico.
```

📝 Os arrays `xs` e `ys` das coordenadas $ x $ e $ y $ dos pontos a serem plotados
devem ter exatamente o mesmo comprimento. Ademais, em geral, quanto maior o
número de pontos especificado, mais preciso será o traço do gráfico. Como exemplo,
experimente diminuir o terceiro argumento de `linspace` na definição de `xs` acima
para $ 6 $ e compare o gráfico produzido.

📝 Se digitarmos apenas `plt.plot(xs, ys)`, sem chamar a função `plt.show()`, o
gráfico não será exibido na maioria dos casos. Entretanto, em alguns ambientes
interativos, como o próprio Jupyter Notebook, a exibição do gráfico pode ser
automática mesmo sem a chamada explícita de `plt.show()`.  Isto ocorre porque o
Jupyter possui um mecanismo integrado para lidar com a renderização de figuras
do Matplotlib. Para habilitar a renderização automática no Jupyter, utilize o
comando mágico `%matplotlib inline` ou `%matplotlib notebook` antes de importar
a Matplotlib.

### $ 2.2 $ Desenhando o gráfico de duas ou mais funções num mesmo diagrama

Em muitas ocasiões é necessário exibir simultaneamente o gráfico de mais de uma
função num mesmo diagrama. Como antes, começamos importanto a Matplotlib e
o NumPy e definindo as nossas funções, por exemplo:
$$
f(x) = \frac{x^3\cos x}{10} + 2x^2 \sin x - 1 \quad \text{e} \quad g(x) = x^2\sin(2x) + 3x - 2\,.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos
f = lambda x: x**3 * cos(x) / 10 + 2 * x**2 * sin(x) - 1
g = lambda x: x**2 * sin(2 * x) + 3 * x - 2
```

Agora precisamos criar uma sequência dos valores $ x $ a serem plotados.
Como antes, isto pode ser feito com ajuda da função `numpy.linspace`. Além
disto, precisamos gerar as seqüências das coordenadas $ y $ dos pontos
sobre os gráficos de $ f $ e $ g $ correspondentes a estes valores de $ x $:

```python
# Gerando um array com 101 pontos igualmente espaçados em [-3, 3]:
xs = np.linspace(-3, 3, 101)
ys_f = f(xs)
ys_g = g(xs)
```

Agora utilizamos a função `matplotlib.pyplot.plot` para criar os plots
correspondentes aos gráficos de $ f $ e de $ g $. Podemos ainda personalizar a
cor e o estilo de linha para cada função, se desejarmos.

```python
# Criando o plot do gráfico de f:
plt.plot(xs, ys_f, label="$ y = f(x) $",
         color="blue", linestyle="-")
plt.plot(xs, ys_g, label="$ y = g(x) $",
         color="red", linestyle="--")

plt.xlabel("eixo-x")    # Especificando o rótulo de eixo-x.
plt.ylabel("eixo-y")    # Especificando o rótulo do eixo-y.
# Especificando um título para o gráfico:
plt.title("Gráfico das funções $ f(x) $ e $ g(x) $")
plt.grid(True)          # Cria um grid retangular.
# Adicionando uma legenda para distingüir as duas funções:
plt.legend()            
plt.show()              # Exibe o gráfico.
```

## $ \S 3 $ Desenhando gráficos de funções de duas variáveis

O procedimento para visualização do gráfico de uma função de duas variáveis é
muito parecido com o descrito na $ \S 1 $. Novamente, começamos importanto o
NumPy e a Matplotlib e definindo a função $ f $ sob consideração, digamos:
$$
f(x, y) = x^2 + y^2\,.
$$

```python
# Importando as bibliotecas e módulos necessários:
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Definindo a função f(x, y) que será plotada:
def f(x, y):
    return x**2 + y**2
```

Primeiramente precisamos gerar dois arrays, `xs` e `ys`, não necessariamente de
mesmo comprimento, contendo as coordenadas-$ x $ e $ y $ dos pontos do domínio
de $ f $ a serem plotados. Para gerá-los, utilizamos novamente a função
`numpy.linspace`. Porém agora precisaremos também gerar uma grade retangular
consistindo de todos os possíveis _pares_ $ (x, y) $ para $ x $ em `xs` e $ y $
em `ys`. E para cada um destes, precisamos avaliar a função $ f $ aí.

```python
xs = np.linspace(-3, 3, 100)    # Gera o array das coordenadas-x.
ys = np.linspace(-3, 3, 200)    # Gera o array das coordenadas-y.
# Gerando duas matrizes X e Y de dimensões 100 x 200 cada uma.
# Cada a linha de X é igual ao array xs,
# e cada linha de Y é uma cópia de ys.
X, Y = np.meshgrid(xs, ys)         
# Finalmente, geramos uma matriz Z cuja (i, j)-ésima entrada é
# f(xs[i], ys[j]):
Z = f(X, Y) 
```

Agora criamos uma figura e um conjunto de eixos 3D usando
`matplotlib.pyplot.subplots` e o módulo `Axes3D` que foi
importado no início:

```python
# Criando a figura a ser plotada no gráfico 3D:
fig = plt.figure()

# Adicionando um conjunto de eixos 3D à figura:
ax = fig.add_subplot(111, projection='3d')

# Plotando a superfície 3D usando os valores em X, Y e Z:
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
# 'viridis' é uma paleta de cores que vai do azul ao amarelo.
# 'alpha=0.8' é usado para especificar a transparência da superfície.

# Adicionando rótulos aos eixos-x, y e z:
ax.set_xlabel('eixo-x')
ax.set_ylabel('eixo-y')
ax.set_zlabel('eixo-z')

# Adicionando um título ao gráfico:
ax.set_title('Gráfico da função $ f(x, y) = x^2 + y^2 $')

# Exibindo o gráfico 3D:
plt.show()
```

Aqui está o código completo utilizado neste exemplo:

```python
# Importando as bibliotecas e módulos necessários:
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Definindo a função f(x, y) que será plotada:
def f(x, y):
    return x**2 + y**2

xs = np.linspace(-3, 3, 100)    # Gera o array das coordenadas-x.
ys = np.linspace(-3, 3, 200)    # Gera o array das coordenadas-y.

# Gerando duas matrizes X e Y de dimensões 100 x 200 cada uma, onde
# cada linha de X (resp. Y) é igual ao array xs (resp. ys):
X, Y = np.meshgrid(xs, ys)         
# Gerando uma matriz Z cuja (i, j)-ésima entrada é f(xs[i], ys[j]):
Z = f(X, Y) 

# Criando a figura a ser plotada no gráfico 3D:
fig = plt.figure()

# Adicionando um conjunto de eixos 3D à figura:
ax = fig.add_subplot(111, projection='3d')

# Plotando a superfície 3D usando os valores em X, Y e Z:
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
# 'viridis' é uma paleta de cores que vai do azul ao amarelo.
# 'alpha=0.8' é usado para especificar a transparência da superfície.

# Adicionando rótulos aos eixos-x, y e z:
ax.set_xlabel('eixo-x')
ax.set_ylabel('eixo-y')
ax.set_zlabel('eixo-z')

# Adicionando um título ao gráfico:
ax.set_title('Gráfico da função $ f(x, y) = x^2 + y^2 $')

# Exibindo o gráfico 3D:
plt.show()
```

## $ \S 4 $ Visualizando gráficos de dispersão com a Matplotlib

Criar um gráfico de dispersão (scatter plot) usando Matplotlib é bastante simples.

```python
# Definindo a função f(x, y) que será plotada:
import numpy as np
import matplotlib.pyplot as plt

# Gerando um array das coordenadas-x e y de maneira aleatória:
xs = np.random.rand(50)
ys = np.random.rand(50)

plt.scatter(xs, ys)
plt.xlabel('eixo-x')
plt.ylabel('eixo-y')
plt.title('Gráfico de dispersão')
plt.show()
```

📝 Como qualquer outra biblioteca, o gerador de números aleatórios
`numpy.random.rand` utiliza um algoritmo _determinístico_. Portanto, embora os
números gerados pareçam aleatórios, na verdade eles são baseados em um valor
inicial, chamado de __semente__ (__seed__). 

O valor padrão da semente usada pelo NumPy é baseado no relógio do sistema, o
que significa que se ela não for definida explicitamente, uma semente diferente
será usada a cada vez que o código for executado. Isto pode ser útil se 
pretendemos gerar conjuntos diferentes de números aleatórios.

Por outro lado, se fornecermos a mesma semente, cada chamada da função `rand`
produzirá exatamente a mesma sequência de números aleatórios. Para definir o
valor da semente utilizamos a função `numpy.random.seed`. Isto garante que a
seqüência de números aleatórios gerada seja reprodutível, o que é útil em
situações em que se deseja obter resultados consistentes e replicáveis.
Experimente adicionar a instrução `np.random.seed(123)` à célula de código acima
(antes das chamadas a `rand`) e verifique que o mesmo gráfico de dispersão é
produzido a cada execução.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

iris = sns.load_dataset('iris')

# Criando um mapa de cor para as diferentes espécies:
species = iris['species'].unique()
species_colors = {sp: i for i, sp in enumerate(species)}
colors = iris['species'].map(species_colors)

plt.scatter(iris['petal_length'], iris['petal_width'], c=colors, cmap='viridis', alpha=0.8)
plt.xlabel('Comprimento da Pétala (cm)')
plt.ylabel('Largura da Pétala (cm)')
plt.title('Comprimento vs. Largura da Pétala das Flores de Íris')

# Criando a legenda:
legend_elements = [mpatches.Patch(color=plt.cm.viridis(i / (len(species) - 1)), label=sp) for i, sp in enumerate(species)]
plt.legend(handles=legend_elements, title='Espécies', loc='upper left')

plt.show()
```

# $ \S 4 $ Criando histogramas

Um __histograma__ é um gráfico que representa a distribuição de freqüência de um
conjunto de dados. Vamos criar um histograma simples usando dados gerados aleatoriamente,
como na $ \S 3 $:

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(14)
dados = np.random.randn(1000)

# Gerando o histograma:
plt.hist(dados, bins=30)
plt.xlabel('Valores')       # Etiquetando o eixo-x.
plt.ylabel('Frequência')    # Etiquetando o eixo-y.
plt.title('Histograma de dados aleatórios seguindo aprox. a distribuição normal')     # Escolhendo um título.
plt.show()
```

📝 A função `numpy.random.randn` gera números aleatórios a partir de uma
_distribuição normal_ padrão (também conhecida como _distribuição Gaussiana_),
cujo valor esperado (média) é igual a $ 0 $ e cujo desvio padrão vale $ 1 $.
`randn` tem por parâmetros as dimensões do array que se deseja criar.
Por exemplo, se você deseja gerar um array 1D com 10 números aleatórios, você
pode usar a função da seguinte maneira: `randn(10)`. Já para criar um array
2D de dimensões $ 3 \times 4 $, utilizaríamos a instrução `randn(3, 4)`.
